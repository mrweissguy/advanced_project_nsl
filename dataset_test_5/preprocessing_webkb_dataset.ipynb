### Due to the lack of data we concatinate alle the data together. While doing this we assume that the content id comparable between universities..

```python
import pandas as pd
```

```python
wisconsin_content = pd.read_csv("WebKB/wisconsin.content", sep="	", header=None)
cornell_content = pd.read_csv("WebKB/cornell.content", sep="	", header=None)
texas_content = pd.read_csv("WebKB/texas.content", sep="	", header=None)
washington_content = pd.read_csv("WebKB/washington.content", sep="	", header=None)
```

```python
# adding the dataframes together.

new_df = pd.concat([wisconsin_content, cornell_content, texas_content, washington_content], ignore_index=True)
```

```python
new_df = new_df.sample(frac=1)
```

```python
new_df.to_csv("collected.content", sep="\t", header = False,index=False)
```

```python
wisconsin_cites = pd.read_csv("WebKB/wisconsin.cites", sep=" ", header=None)
cornell_cites = pd.read_csv("WebKB/cornell.cites", sep=" ", header=None)
texas_cites = pd.read_csv("WebKB/texas.cites", sep=" ", header=None)
washington_cites = pd.read_csv("WebKB/washington.cites", sep=" ", header=None)
```

```python
new_df_cites = pd.concat([wisconsin_cites, cornell_cites, texas_cites, washington_cites], ignore_index=True)
```

```python
new_df_cites = new_df_cites.sample(frac=1)
```

```python
new_df_cites.to_csv("collected.cites", sep="\t", header = False,index=False)
```
